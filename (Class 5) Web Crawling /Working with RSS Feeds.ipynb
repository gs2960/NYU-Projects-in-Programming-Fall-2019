```python
import feedparser
import datetime
from time import sleep
import pandas as pd
from pandas import DataFrame
from re import match, IGNORECASE
from random import choice
```

```python
nytimes_rss_url = 'http://rss.nytimes.com/services/xml/rss/nyt/HomePage.xml'
nytimes_feed = feedparser.parse(nytimes_rss_url)

nytimes_feed
```

```python
nytimes_title_list = []
for i in range(30): 
    nytimes_story = nytimes_feed['entries'][i]['title']
    nytimes_title_list.append(nytimes_story)
    
nytimes_title_list[0] # the latest article in the RSS feed 
```

```python
nytimes_link_list = []
for i in range(30):
    nytimes_story = nytimes_feed['entries'][i]['link']
    nytimes_link_list.append(nytimes_story)
    
nytimes_link_list[0]
```

```python
nytimes_description_list = []
for i in range(30):
    nytimes_story = nytimes_feed['entries'][i]['description']
    nytimes_description_list.append(nytimes_story)
    
nytimes_description_list[0]
```

```python
for i in nytimes_title_list:
    df = DataFrame({"article_title":nytimes_title_list,"description":nytimes_description_list,"link":nytimes_link_list})

df = df[["article_title", "description", "link"]]

df.head()
```
